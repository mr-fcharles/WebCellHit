```python
import pandas as pd
import numpy as np
from pathlib import Path
from tqdm.auto import tqdm

import sys
sys.path.append('/home/fcarli/CellHit/')
from CellHit.utils import QuantileScoreComputer
```

```python
#specific dataset and results directory
dataset = 'gdsc'
path = Path(f'/home/fcarli/WebCellHit/webserver_data/local_data/inference_results/{dataset}')
```

```python
dfs = []

for file in tqdm(path.glob('*.csv')):
    df = pd.read_csv(file,index_col=0)
    dfs.append(df)

#assemble the dataframes
assembled_df = pd.concat(dfs,ignore_index=True)
assembled_df
```

```python
quantile_computer = QuantileScoreComputer(assembled_df,cell_col='index',drug_col='DrugID',score_col='prediction')
#save the computer
quantile_computer.save(f'/home/fcarli/WebCellHit/webserver_data/local_data/{dataset}_quantile_computer.npy')
```

```python
assembled_df['QuantileScore'] = quantile_computer.parallel_compute_score(assembled_df,cell_col='index',drug_col='DrugID',score_col='prediction',n_jobs=26)
```
