# Precomptued transcriptomics KNN

```python
#read alligner data
transcriptomics = pd.read_feather('/home/fcarli/WebCellHit/data/transcriptomics/celligner_CCLE_TCGA_optimized_revised.feather')
```

```python
transcriptomics
#keep only rows with "ACH-" and "TCGA-" in the index
transcriptomics = transcriptomics[transcriptomics['index'].str.contains('ACH-|TCGA-')]
```

```python
ccle_trascr = transcriptomics[transcriptomics['Source'] == 'CCLE'].set_index('index').drop('Source',axis=1)
tcga_trascr = transcriptomics[transcriptomics['Source'] == 'TCGA'].set_index('index').drop('Source',axis=1)
```

```python
ccle_trascr
```

```python
ccle_neighs = FaissKNN()
ccle_neighs.fit(ccle_trascr)

tcga_neighs = FaissKNN()
tcga_neighs.fit(tcga_trascr)

ccle_neighs.save('/home/fcarli/WebCellHit/webserver_data/local_data/ccle_transcr_neighs.pkl')
tcga_neighs.save('/home/fcarli/WebCellHit/webserver_data/local_data/tcga_transcr_neighs.pkl')
```

```python
#load ccle neighs
ccle_neighs = FaissKNN.from_file('/home/fcarli/WebCellHit/webserver_data/local_data/ccle_transcr_neighs.pkl')
tcga_neighs = FaissKNN.from_file('/home/fcarli/WebCellHit/webserver_data/local_data/tcga_transcr_neighs.pkl')
```

```python
ccle_neighs_df = ccle_neighs.knn(ccle_trascr.sample(10),k=4,skip_self=True)
tcga_neighs_df = tcga_neighs.knn(tcga_trascr.sample(10),k=4,skip_self=True)
```
